```python
# Importing Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
from requests.exceptions import ConnectionError
```

```python
# Entering job keyword & country
job_keyword = "sales"
country = "United States"

# creating search url with job_key & location embeded in it
base_url = 'https://www.linkedin.com/jobs/search'
payload = {"keywords":job_keyword,"location":country}

# get() API
try:
    response  = requests.get(base_url,params=payload)
    
# handling connection errors
except ConnectionError as e:
    print(e)

else:
    # check if we get OK request
    if response.status_code == 200 :
        # assigning html content to a variable
        html_content = response.content
        # creating Beautiful Soup Object
        soup = BeautifulSoup(html_content,"html.parser")
        # find the search url (optional)
        search_tag = soup.find("link")
        search_link = search_tag.get("href")
        print(search_link)
    else:
        print(response)
```

```python
# find all job links with find_all() method
# we want to find all <a> tags with specific class attr
job_tags = soup.find_all("a",{"class":"base-card__full-link absolute top-0 right-0 bottom-0 left-0 p-0 z-[2]"})
# dispaly the jobs list
job_tags
```

```python
# store all job links in a list
job_links = []
for tag in job_tags:
    job_links.append(tag.get("href"))

job_links
```

```python
# extract job titles with .text method
# store all job titles in a list
job_titles = []
for tag in job_tags:
    job_titles.append(tag.text.strip())

job_titles
```

```python
# creating pandas data frame
vacancy_dict = {"Vacancy":job_titles,"Apply Link":job_links}
jobs_df = pd.DataFrame(vacancy_dict)
jobs_df
```

```python
# Function to make clickable links
def make_clickable(url, name):
    return '<a href="{}" rel="noopener noreferrer" target="_blank">{}</a>'.format(url,name)

# adding a new column to the data frame that combines the other two columns in one clickable column
jobs_df['Job Vacancy'] = jobs_df.apply(lambda x: make_clickable(x['Apply Link'], x['Vacancy']), axis=1)
```

```python
# view the data frame with clickable links
jobs_df[["Job Vacancy"]].style
```

```python
# find all company links with find_all() method
company_tags = soup.find_all("a",{"class":"hidden-nested-link"})
company_tags
```

```python
# store all company links in a list
company_links = []
for company in company_tags:
    company_links.append(company.get("href"))

company_links
```

```python
# extract job titles with .text method
# store all job titles in a list
company_names = []
for company in company_tags:
    company_names.append(company.text.strip())

company_names
```

```python
# creating pandas data frame
company_dict = {"Company":company_names,"Company Profile":company_links}
company_df = pd.DataFrame(company_dict)
company_df
```

```python
# Function to make clickable links
def make_clickable(url, name):
    return '<a href="{}" rel="noopener noreferrer" target="_blank">{}</a>'.format(url,name)

# adding a new column to the data frame that combines the other two columns in one clickable column
company_df['Company Name'] = company_df.apply(lambda x: make_clickable(x['Company Profile'], x['Company']), axis=1)
```

```python
# display the company Data Frame with clickable links
company_df[['Company Name']].style
```

```python
# joining Jobs Data Frame with Company Data Frame
jobs_df[["Job Vacancy"]].join(company_df[['Company Name']]).style
```

```python
# assigning file path to wrtie in to it
# I chose to creat the file to my home directory
# feel free to add a path which suites you
file_path = "LinkedIn_Jobs_{}_{}.html".format(job_keyword,country)
# joining Jobs Data Frame with Company Data Frame
jobs_tbl_df = jobs_df[["Job Vacancy"]].join(company_df[['Company Name']]).style
# exporting Data Frame to HTML document using to_html() method
jobs_tbl_df.to_html(file_path)
```
